### Assignment 2.4.1
For the given example, implement a linear regression minimizing the sum of squared
error from scratch.
Please provide a plot containing the original data and your fitted line.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("linreg_data.csv")

X = df['X'].values
Y = df['Y'].values
mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)

m = np.sum((X - mean_x) * (Y - mean_y)) / np.sum((X - mean_x) ** 2)
b = mean_y - (m * mean_x)

Y_prediction = m * X + b

plt.scatter(X, Y, label="Data")
plt.plot(X, Y_prediction, color='r', label="Fitted line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

### Assignment 2.4.2
Do the same task again, but this time using "scipy.stats.linregress".
Again Provide a plot containing the original data and your fitted line.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

df = pd.read_csv("linreg_data.csv")

X = df['X'].values
Y = df['Y'].values

slope, intercept, _, _, _ = linregress(X, Y)
Y_prediction = slope * X + intercept

plt.scatter(X, Y, label="Data")
plt.plot(X, Y_prediction, color='r', label="Fitted line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

### Assignment 2.4.3
Use this regression to fill in a missing value at X = 2. What is the corresponding
Y value?

```python
X_value = 2
Y_value = slope * X_value + intercept
print(f"The corresponding value is Y_{X_value} = {Y_value}.")
```

### Assignment 2.4.4
How can we use the regression to smooth the original observations?

We can replace the Y values in the dataframe with the predicted values from the regression to smooth the observations out.

```python
smoothed_values = np.array(Y_prediction)

plt.scatter(X, smoothed_values, label="Data")
plt.plot(X, Y_prediction, color='r', label="Fitted line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
